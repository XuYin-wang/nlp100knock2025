
61. 特徴ベクトル
Bag of Words (BoW) に基づき、学習データ（train.tsv）および検証データ（dev.tsv）のテキストを特徴ベクトルに変換したい。ここで、ある事例のテキストの特徴ベクトルは、テキスト中に含まれる単語（スペース区切りのトークン）の出現頻度で構成する。例えば、”too loud , too goofy”というテキストに対応する特徴ベクトルは、以下のような辞書オブジェクトで表現される。

{'too': 2, 'loud': 1, ',': 1, 'goofy': 1}

各事例はテキスト、特徴ベクトル、ラベルを格納した辞書オブジェクトでまとめておく。例えば、先ほどの”too loud , too goofy”に対してラベル”0”（ネガティブ）が付与された事例は、以下のオブジェクトで表現される。

{'text': 'too loud , too goofy', 'label': '0', 'feature': {'too': 2, 'loud': 1, ',': 1, 'goofy': 1}}

学習データと検証データの各事例を上記のような辞書オブジェクトに変換したうえで、学習データと検証データのそれぞれを、辞書オブジェクトのリストとして表現せよ。さらに、学習データの最初の事例について、正しく特徴ベクトルに変換できたか、目視で確認せよ。




```python
from collections import Counter
import os
import pandas as pd

# BoW 特徴関数
def text_to_bow(text):
    tokens = text.split()          # スペースでトークン化
    return dict(Counter(tokens))   # 単語の出現頻度をカウント

# データ変換関数
# 入力を data_df とすることで、DataFrameであることを明確にする
def convert_dataset(data_df):
    dataset = []
    for _, row in data_df.iterrows():
        example = {
            'text': row['sentence'],
            'label': row['label'],
            'feature': text_to_bow(row['sentence'])
        }
        dataset.append(example)
    return dataset

# ここでTSVファイルを読み込みます。
reloaded_train_df = pd.read_csv(os.path.join(path, 'train.tsv'), sep='\t')
reloaded_dev_df = pd.read_csv(os.path.join(path, 'dev.tsv'), sep='\t')

train_data = convert_dataset(reloaded_train_df)
dev_data = convert_dataset(reloaded_dev_df)

# .json として保存する処理
import json

with open('train_data.json', 'w') as f:
    json.dump(train_data, f, indent=2)

with open('dev_data.json', 'w') as f:
    json.dump(dev_data, f, indent=2)
```

62. 学習
61で構築した学習データの特徴ベクトルを用いて、ロジスティック回帰モデルを学習せよ。




```python
import json

from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression

# 学習データの読み込み
# train_data.json は json 形式で保存されているため、json.load() を使用する
with open('train_data.json', 'r') as f: # テキストモードで開くため 'r' に変更
    train_data_loaded = json.load(f)

# 特徴量 (X) とラベル (y) を抽出
X_train = [d['feature'] for d in train_data_loaded]
y_train = [int(d['label']) for d in train_data_loaded]

vectorizer = DictVectorizer(sparse=True)
X_train_vec = vectorizer.fit_transform(X_train)

# ロジスティック回帰モデルの学習
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)
```

63. 予測
学習したロジスティック回帰モデルを用い、検証データの先頭の事例のラベル（ポジネガ）を予測せよ。また、予測されたラベルが検証データで付与されていたラベルと一致しているか、確認せよ。



```python
# 検証データの読み込み
with open('train_data.json', 'r') as f:
    dev_data_loaded = json.load(f)

#予測するデータ
dev_example = dev_data_loaded[20]

X_dev = [dev_example['feature']]
y_true = int(dev_example['label'])

X_dev_vec = vectorizer.transform(X_dev)
y_pred = model.predict(X_dev_vec)[0]

print("テキスト:", dev_example['text'])
print("予測ラベル:", y_pred)
print("正解ラベル:", y_true)

if y_pred == y_true:
    print("予測は正解です")
else:
    print("予測は不正解です")
```

64. 条件付き確率
学習したロジスティック回帰モデルを用い、検証データの先頭の事例を各ラベル（ポジネガ）に分類するときの条件付き確率を求めよ。



```python
import json
import pickle

# 検証データの読み込み
# dev_data.json は json 形式で保存されているため、json.load() を使用する
with open('dev_data.json', 'r') as f:
    dev_data_loaded = json.load(f)

# 予測する検証データの先頭の事例
dev_example_first = dev_data_loaded[0]

# 特徴量 (X) を抽出し、ベクトル化
X_dev_first = [dev_example_first['feature']]
X_dev_first_vec = vectorizer.transform(X_dev_first)

# 条件付き確率の計算
# model.predict_proba() は各クラスの確率を返します。
probabilities = model.predict_proba(X_dev_first_vec)[0]

print("検証データの最初の事例のテキスト:", dev_example_first['text'])
print("正解ラベル:", dev_example_first['label'])
print("各ラベルに対する条件付き確率:")
# クラスの順序は model.classes_ で確認できます
for i, prob in enumerate(probabilities):
    print(f"  ラベル {model.classes_[i]}: {prob:.4f}")
```

65. テキストのポジネガの予測
与えられたテキストのポジネガを予測するプログラムを実装せよ。例えば、テキストとして”the worst movie I ‘ve ever seen”を与え、ロジスティック回帰モデルの予測結果を確認せよ。



```python
from collections import Counter

def text_to_bow(text):
    tokens = text.split()
    return dict(Counter(tokens))

def predict_sentiment(text, model, vectorizer):
    feature = text_to_bow(text)
    X_vec = vectorizer.transform([feature])

    pred_label = model.predict(X_vec)[0]
    prob = model.predict_proba(X_vec)[0]

    return pred_label, prob
test = "the worst movie I 've ever seen"

label, prob = predict_sentiment(test, model, vectorizer)

print("テキスト:", test)
print("予測ラベル:", label)
print("P(ネガティブ):", prob[0])
print("P(ポジティブ):", prob[1])
```

66. 混同行列の作成
学習したロジスティック回帰モデルの検証データにおける混同行列（confusion matrix）を求めよ。

```python
# 検証データの特徴とラベル
X_dev = [d['feature'] for d in dev_data_loaded]
y_dev = [int(d['label']) for d in dev_data_loaded]

X_dev_vec = vectorizer.transform(X_dev)
y_pred = model.predict(X_dev_vec)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_dev, y_pred)
print("混同行列:")
print(cm)
```



67. 精度の計測
学習したロジスティック回帰モデルの正解率、適合率、再現率、F1スコアを、学習データおよび検証データ上で計測せよ。

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# 学習データ
X_train = [d['feature'] for d in train_data_loaded]
y_train = [int(d['label']) for d in train_data_loaded]

X_train_vec = vectorizer.transform(X_train)
y_train_pred = model.predict(X_train_vec)

print("Train Data")
print("Accuracy :", accuracy_score(y_train, y_train_pred))
print("Precision:", precision_score(y_train, y_train_pred))
print("Recall   :", recall_score(y_train, y_train_pred))
print("F1-score :", f1_score(y_train, y_train_pred))
# 検証データ
X_dev = [d['feature'] for d in dev_data_loaded]
y_dev = [int(d['label']) for d in dev_data_loaded]

X_dev_vec = vectorizer.transform(X_dev)
y_dev_pred = model.predict(X_dev_vec)

print("\n Dev Data")
print("Accuracy :", accuracy_score(y_dev, y_dev_pred))
print("Precision:", precision_score(y_dev, y_dev_pred))
print("Recall   :", recall_score(y_dev, y_dev_pred))
print("F1-score :", f1_score(y_dev, y_dev_pred))
```



68. 特徴量の重みの確認
学習したロジスティック回帰モデルの中で、重みの高い特徴量トップ20と、重みの低い特徴量トップ20を確認せよ。



```python
import numpy as np

# ロジスティック回帰の重み
weights = model.coef_[0]

# 特徴量（単語）名
feature_names = vectorizer.get_feature_names_out()
feature_weights = list(zip(feature_names, weights))
top_positive = sorted(
    feature_weights,
    key=lambda x: x[1],
    reverse=True
)[:20]

print("重みの高い特徴量 Top 20")
for word, weight in top_positive:
    print(f"{word}: {weight:.4f}")
top_negative = sorted(
    feature_weights,
    key=lambda x: x[1]
)[:20]

print("\n 重みの低い特徴量 Top 20")
for word, weight in top_negative:
    print(f"{word}: {weight:.4f}")
```

69. 正則化パラメータの変更
ロジスティック回帰モデルを学習するとき、正則化の係数（ハイパーパラメータ）を調整することで、学習時の適合度合いを制御できる。正則化の係数を変化させながらロジスティック回帰モデルを学習し、検証データ上の正解率を求めよ。実験の結果は、正則化パラメータを横軸、正解率を縦軸としたグラフにまとめよ。

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction import DictVectorizer

# データ読み込み
with open('train_data.json', 'r') as f:
    train_data_loaded = json.load(f)

with open('dev_data.json', 'r') as f:
    dev_data_loaded = json.load(f)

# 特徴量とラベル
X_train = [d['feature'] for d in train_data_loaded]
y_train = [int(d['label']) for d in train_data_loaded]

X_dev = [d['feature'] for d in dev_data_loaded]
y_dev = [int(d['label']) for d in dev_data_loaded]

# ベクトル化（train データのみで fit させる）
vectorizer = DictVectorizer(sparse=True)
X_train_vec = vectorizer.fit_transform(X_train)
X_dev_vec = vectorizer.transform(X_dev)

# 正則化パラメータ
C_values = np.logspace(-3, 2, 10)
accuracies = []

# 学習と評価
for C in C_values:
    model = LogisticRegression(
        C=C,
        max_iter=1000,
        solver='liblinear'
    )
    model.fit(X_train_vec, y_train)
    y_pred = model.predict(X_dev_vec)
    acc = accuracy_score(y_dev, y_pred)
    accuracies.append(acc)

# グラフ描画
plt.figure()
plt.plot(C_values, accuracies, marker='o')
plt.xscale('log')
plt.xlabel("正則化パラメータ C")
plt.ylabel("検証データ上の正解率")
plt.title("ロジスティック回帰における正則化の効果")
plt.show()
```
