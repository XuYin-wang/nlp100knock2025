42. 多肢選択問題の正解率


JMMLU のいずれかの科目を大規模言語モデルに解答させ、その正解率を求めよ。

```python
import google.generativeai as genai

genai.configure(api_key = "AIzaSyDjmdhC5cGQ1J3klCZ8wfXqVM3zwDL2KAg")

# 1. 問題定義（JMMLU形式の多肢選択問題）
math_questions = [
    {
        "question": """以下の選択肢のうち、絶対値が1となるのはどれですか？
A. 1
B. -1
C. 2
D. -2

答えの選択肢をカンマ区切りで回答してください。""",
        "answer": {"A", "B"}  # 正解の選択肢セット
    },
    {
        "question": """以下の選択肢のうち、偶数はどれですか？
A. 1
B. 2
C. 3
D. 4

答えの選択肢をカンマ区切りで回答してください。""",
        "answer": {"B", "D"}
    }
]

# 2. Geminiモデルの初期化
gemini_model = genai.GenerativeModel('gemini-pro-latest')

# 3. モデルに質問を提示し、回答を抽出する関数
def get_model_answer(question: str) -> set:
    response = gemini_model.generate_content(question)
    # モデル回答から 'A'～'D' の文字を抽出して集合にする
    return {ch for ch in response.text.upper() if ch in {"A", "B", "C", "D"}}

# 4. モデル回答の評価
for i, item in enumerate(math_questions, start=1):
    model_response = get_model_answer(item["question"])  # モデル回答取得
    correct_answer_set = item["answer"]  # 正解セット
    # 正解率計算（正しく識別された選択肢数 / 正解選択肢数）
    correct_rate = len(model_response & correct_answer_set) / len(correct_answer_set)

    # 結果出力
    print(f"\n--- 問題 {i} ---")
    print(f"問題:\n{item['question']}")
    print(f"モデルの回答: {model_response}")
    print(f"正解: {correct_answer_set}")
    print(f"正解率: {correct_rate:.2f}")
```

43. 応答のバイアス


問題42において、実験設定を変化させると正解率が変化するかどうかを調べよ。実験設定の例としては、大規模言語モデルの温度パラメータ、プロンプト、多肢選択肢の順番、多肢選択肢の記号などが考えられる。
正解の選択肢を全てDに入れ替えて解答させる例。

```python
import google.generativeai as genai

genai.configure(api_key = "AIzaSyDjmdhC5cGQ1J3klCZ8wfXqVM3zwDL2KAg")

# 問題定義（正解は本来 A,Bなど）
math_questions = [
    {
        "question": "絶対値が1となるのはどれですか？\nA. 1\nB. -1\nC. 2\nD. -2",
        "answer": {"A", "B"}
    }
]

gemini_model = genai.GenerativeModel('gemini-pro-latest')

def get_forced_d_answer(question: str) -> set:
    """
    モデルに「全ての正解は D」と強制するプロンプトで回答を取得
    """
    forced_prompt = question + "\n※全ての正解は D に置き換えて回答してください。"
    response = gemini_model.generate_content(forced_prompt)
    # モデル回答から D を抽出
    return {ch for ch in response.text.upper() if ch == "D"}

for i, item in enumerate(math_questions, start=1):
    model_response = get_forced_d_answer(item["question"])
    correct_answer_set = item["answer"]
    correct_rate = len(model_response & correct_answer_set) / len(correct_answer_set)

    print(f"\n--- 問題 {i} ---")
    print(f"問題:\n{item['question']}")
    print(f"モデルの回答（全て D に置き換え）: {model_response}")
    print(f"正解: {correct_answer_set}")
    print(f"正解率: {correct_rate:.2f}")
```

つばめちゃんは渋谷駅から東急東横線に乗り、自由が丘駅で乗り換えました。東急大井町線の大井町方面の電車に乗り換えたとき、各駅停車に乗車すべきところ、間違えて急行に乗車してしまったことに気付きました。自由が丘の次の急行停車駅で降車し、反対方向の電車で一駅戻った駅がつばめちゃんの目的地でした。目的地の駅の名前を答えてください。


```python
# 東急大井町線の駅順序（自由が丘→大井町方面）
stations_order = [
    "自由が丘", "九品仏", "尾山台", "等々力", "上野毛", "二子玉川",
    "二子新地", "高津", "溝の口", "梶が谷", "宮前平", "鷺沼",
    "たまプラーザ", "あざみ野", "江田", "市が尾", "藤が丘", "青葉台",
    "田奈", "長津田", "恩田", "こどもの国線分岐"
]

# 急行停車駅
express_stops = ["自由が丘", "二子玉川", "溝の口", "大井町"]

# 起点
start_station = "自由が丘"

# 急行の次の停車駅
# 自由が丘のインデックスより大きく、かつ急行停車駅である最初の駅を探す
next_stop = next(station for station in express_stops if stations_order.index(station) > stations_order.index(start_station))

# 反対方向に一駅戻る
# 自由が丘の次の急行停車駅 (二子玉川) で降りたので、目的地はその一駅手前 (上野毛) となる
destination_index = stations_order.index(next_stop) - 1
destination = stations_order[destination_index]

print(f"つばめちゃんの目的地は: {destination}")
```

8. LLMによる評価の頑健性

問題47で行ったLLMによるテキストの評価に関して、その頑健さ（脆弱さ）を調査せよ。最も単純な方法は、同じ評価を何回か繰り返した時のスコアの分散を調べることであろう。また、川柳の末尾に特定のメッセージを追加することで、評価スコアを恣意的に操作することも可能であろう。


```python
import re
import numpy as np
import os
import google.generativeai as genai


genai.configure(api_key = "AIzaSyDjmdhC5cGQ1J3klCZ8wfXqVM3zwDL2KAg")

model = genai.GenerativeModel('gemini-pro-latest')


# 評価
prompt_template = """
次の川柳を「面白さ10段階（1〜10）」で評価し、
必ず「点数のみ」を出力してください。

川柳：
{senryu}
"""

# 振り返し
repeat_times = 3
scores = []

for _ in range(repeat_times):
    prompt = prompt_template.format(senryu=senryu)
    response = model.generate_content(prompt)
    text = response.text.strip()

    # 数字
    match = re.search(r"\d+", text)
    if match:
        scores.append(int(match.group()))

# 計算する
mean = np.mean(scores)
std = np.std(scores)
var = np.var(scores)

print("評価スコア:", scores)
print(f"平均: {mean:.2f}, 標準偏差: {std:.2f}, 分散: {var:.2f}")
```
